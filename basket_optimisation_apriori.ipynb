```python
# 分析basket optimization 中的频繁项集和关联规则
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

# 数据加载
mbo=pd.read_csv("Market_Basket_Optimisation.csv",header=None)
mbo.fillna("null",inplace=True)
temp_df=pd.DataFrame(mbo[0]+'|'+mbo[1]+'|'+mbo[2]+'|'+mbo[3]+'|'+mbo[4]+'|'+mbo[5]+'|'+mbo[6]+'|'+mbo[7]+'|'+mbo[8]+'|'+mbo[9]+'|'+mbo[10]+'|'+mbo[11]+'|'+mbo[12]+'|'+mbo[13]+'|'+mbo[14]+'|'+mbo[15]+'|'+mbo[16]+'|'+mbo[17]+'|'+mbo[18]+'|'+mbo[19])
temp_df.columns=['trans']
temp_df['trans_new']=temp_df['trans'].map(lambda x:x.replace('|null',''))
print("----:",temp_df['trans_new'])
hot_encoded=temp_df.drop(['trans_new','trans'],1).join(temp_df.trans_new.str.get_dummies(sep='|'))
pd.options.display.max_columns=100
print(hot_encoded.head(1))


# 挖掘频繁项集，最小支持度为0.04
itemsets = apriori(hot_encoded,use_colnames=True, min_support=0.04)
# 按照支持度从大到小进行时候粗
itemsets = itemsets.sort_values(by="support" , ascending=False) 
print('-'*20, '频繁项集', '-'*20)
print(itemsets)
# 根据频繁项集计算关联规则，设置最小提升度为1.5
rules =  association_rules(itemsets, metric='lift', min_threshold=1.5)
# 按照提升度从大到小进行排序
rules = rules.sort_values(by="lift" , ascending=False) 
#rules.to_csv('./rules.csv')
print('-'*20, '关联规则', '-'*20)
print(rules)
```
